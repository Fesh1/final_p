

We can see the SpType contains the Roman Numerals we need.

    Best Practice is to use Regex (Regular Expression) to extract the pattern.

But I will try to use an intuitive approach first.

I just took a look from the dataset, the Roman Numeral contains: I, II, III, IV, V, VI, VII in the string. Therefore:

    Dwarfs (I, II, III, VII)
    Giants (IV, V, VI)
    Other Special Stars (None)

^Edit0821: Sorry I made a huge mistake here... Now fixed!

For the First Character of the String, we don't need to worry there will be I or V. 

```python
X['new_target'] = X['SpType']
```

```python
from sys import getsizeof
```

```python
getsizeof('III')
```

```python


#The intuitive approach (Could take a long time if you have a huge dataset)
for i in range(X.shape[0]):

    if "V" in str(X.loc[i,'new_target']): 
        if "VI" in X.loc[i,'new_target']:
            if "VII" in X.loc[i,'new_target']:
                X.loc[i,'new_target'] = 7 # VII is Dwarf
            else:
                X.loc[i,'new_target'] =6 # VI is Giants                
        else:
            if "IV" in X.loc[i,'new_target']: 
                X.loc[i,'new_target'] = 4 # IV is Giants
            elif "V" in X.loc[i,'new_target']: 
                X.loc[i,'new_target'] = 5 # V is Giants

    elif "I" in X.loc[i,'new_target']: 
        if "II" in X.loc[i,'new_target']:
            if "III" in X.loc[i,'new_target']:
                X.loc[i,'new_target'] = 3 # III are Dwarfs
            else:
                X.loc[i,'new_target'] = 2 # II are Dwarfs
        else:
            X.loc[i,'new_target'] = 1 # I are Dwarfs
    else: 
        X.loc[i,'new_target'] = getsizeof(X.loc[i,'new_target']) # None
        

```
